# scatter plot


```python
a=[10,20,30,40,50,60,70,80]
b=[5,2,4,2,1,4,5,2]
x=[1,2,3,4,3,2,1,2]
fig=plt.figure(figsize=(10,5))
ax1=plt.subplot()
ax1.scatter(a,b,c='g',s=200,edgecolors='y',marker='o',alpha=0.5)
ax1.scatter(a,x,c='r',s=400,marker='4',alpha=0.5)
plt.title("Scatter plot")
plt.grid(True)
plt.savefig('ScatterPlot.png')
plt.show()
```

```python
a=[10,20,30,40,50,60,70,80]
b=[5,2,4,2,1,4,5,2]
x=[1,2,3,4,3,2,1,2]
fig=plt.figure(figsize=(10,5))

plt.scatter(a,b)
plt.scatter(a,x)
plt.xlabel("x label")
plt.ylabel("y label")
plt.title("x label y l")
plt.show()
```

```python
a=[10,20,30,40,50,60,70,80]
b=[5,2,4,2,1,4,5,2]
x=[1,2,3,4,3,2,1,2]
fig=plt.figure(figsize=(10,5))
ax1=plt.subplot(1,2,1)
ax1.scatter(a,b,color='r',marker=6,s=200)
plt.title("left")
plt.xlabel("x label 1")
plt.ylabel("y label 1")

ax2=plt.subplot(1,2,2)
ax2.scatter(a,x,color='g',s=200)
plt.title("right")
plt.xlabel('x label 2')
plt.ylabel('y label 2')
plt.show()
```

# Box plot and violin plot

```python
total=[20,4,1,30,10,20,70,30,10]
order=[10,3,1,15,17,2,30,44,2,1]
discount=[30,10,20,5,10,20,50,60,20,45]
data=list([total,order,discount])
fig=plt.figure(figsize=(10,5))
ax1=plt.subplot()
ax1.boxplot(data,showmeans=True)
plt.grid(True)
plt.show()
```
